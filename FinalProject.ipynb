```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('vgsales.csv')
df.head()
```

```python
df['Publisher'].groupby(df['Publisher']).count().sort_values(ascending=False)
```

```python
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)
publisher_sales
```

```python
df['Platform'].groupby(df['Platform']).count().sort_values(ascending=False).head(10)
```

```python
platform_sales = df.groupby('Platform')['Global_Sales'].mean().sort_values(ascending=False)
```

```python
North_America_Sales = df.drop(columns=['EU_Sales','JP_Sales','Other_Sales','Global_Sales'])
```

```python
North_America_Sales = North_America_Sales.drop(North_America_Sales[North_America_Sales['NA_Sales'] == 0].index)
```

I created a new data frame that only has the data for games sold in North America.

```python
North_America_Sales.sort_values(by='NA_Sales', ascending=False).head(10)
```

```python
df['Genre'].groupby(df['Genre']).count().sort_values(ascending=False).head(10)
```

```python
df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
top_50_games = df.sort_values(by='Global_Sales', ascending=False).head(50)
top_50_games
```

```python
plt.subplots(figsize=(8, 13))
sns.barplot(x='Global_Sales', y='Name', data=top_50_games, label = 'Global Sales', color = 'black')
sns.barplot(x='NA_Sales', y='Name', data=top_50_games, label = 'North America Sales', color = 'red')
sns.barplot(x='EU_Sales', y='Name', data=top_50_games, label = 'Europe Sales', color = 'green')
sns.barplot(x='JP_Sales', y='Name', data=top_50_games, label = 'Japan Sales', color = 'yellow')
sns.barplot(x='Other_Sales', y='Name', data=top_50_games, label = 'Other Sales', color = 'blue')
plt.legend(loc='lower right')
plt.xlabel('Sales')
plt.ylabel('Name')
plt.title('Top 50 Games by Global Sales')
plt.show()
```

```python
year_global_sales = df.groupby('Year')['Global_Sales'].sum().sort_values(ascending=False)
year_global_sales.head(10)
```

```python
# Convert the Series to a DataFrame and reset the index
year_global_sales_df = year_global_sales.reset_index()
year_global_sales_df.columns = ['Year', 'Global_Sales']

# Create the scatter plot
sns.scatterplot(x='Year', y='Global_Sales', data=year_global_sales_df)
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Global Sales by Year')
plt.show()
```
